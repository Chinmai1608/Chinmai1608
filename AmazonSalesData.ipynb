```python
sales = pd.read_csv('/content/drive/MyDrive/Budget Sales data/AdventureWorks_Database.xlsx - Sales.csv')
sales = sales.iloc[:,0:13]
sales.head()
```

```python
territory = pd.read_csv('/content/drive/MyDrive/Budget Sales data/AdventureWorks_Database.xlsx - Territory.csv')
territory.head()
```

```python
sales_prod = pd.merge(product,sales,on='ProductKey')
sales_prod.head()
```

```python
sales_prod_cust = pd.merge(sales_prod,customer,on='CustomerKey')
```

```python
sales_prod_cust_terr = pd.merge(sales_prod_cust,territory,on='SalesTerritoryKey')
```
